```python
# JijModeling
import jijzept as jz
import jijmodeling as jm
import json

def sample_problem():
    d = jm.PlaceholderArray("d", dim=1)
    n = d.shape[0]
    x = jm.BinaryArray("x", n)
    i = jm.Element("i", n)
    term = jm.Constraint("const", jm.Sum({i: n}, d[i] * x[i]) == 1)
    term2 = jm.Constraint("const2", jm.Sum({i: n}, d[i] * x[i]) == 1)
    problem = jm.Problem("prob")
    problem += term
    problem += term2
    
    return problem
    

def fast_decode(problem, ph_value, num_reads, num_sweeps):
    client = jz.client.JijZeptClient(config="/home/azureuser/.config/jijzept/config.toml")

    term_obj = jm.to_serializable(problem)

    jijmodel = {
        "mathematical_model": json.dumps(term_obj),
        "instance_data": ph_value
    }

    client.post_instance(instance_type="JijModeling", instance=jijmodel)

    sa_params = {
        "num_reads": num_reads,
        "num_sweeps": num_sweeps,
        "beta_min": None,
        "beta_max": None,
        "updater": None,
    }
    sa_params["multipliers"] = {"const": 0.1}
    sa_params["mul_search"] = True
    client.submit_solve_query(
        queue_name="testsolver",
        solver_name="SAParaSearch",
        parameters=sa_params,
    )
    while True:
        response = client.fetch_result()
        if response["status"] == "SUCCESS":
            break
    print(response["solution"]["info"])
    
    return response
```

```python
import jijbench as jb

target = jb.JijModelingTarget(
    problem=sample_problem(),
    instance=[("sample_instance", {"d": [1, 2, 3, 4]})]
)

bench = jb.Benchmark(
    params={"num_reads": [10], "num_sweeps": [10]},
    solvers=fast_decode,
    targets=target
)
bench.run()
```
