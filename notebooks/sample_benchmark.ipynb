```python
import jijbench as jb
import jijmodeling as jm


def solver1():
    return 100, 1000, 10000

def solver2(qubo, a=2, b=3):
    return f"(x, a, b) = ({qubo}, {a}, {b})"

def gen_model1():
    d1 = jm.Placeholder("d")
    x = jm.Binary("x", shape=(2,))
    problem = jm.Problem("model1")
    problem += x[0] + d1 * x[1]
    return problem

def gen_model2():
    d2 = jm.Placeholder("d")
    x = jm.Binary("x", shape=(2,))
    problem = jm.Problem("model2")
    problem += x[0] + d2 * x[1]
    problem += jm.Constraint("onehot", x[:] == 1)
    return problem


# 今リリースされてるやつでは...
# (instance_name, ph_value)
targets = {
    gen_model1(): [("model1_case1", {"d1": 1}), ("model1_case2", {"d1": 2})],
    gen_model2(): [("model2_case1", {"d2": 10})],
}

# 以下に変更予定
#problemはjijmodelingのproblem
#nstanceは[(instance_name, ph_value),...] 
t1 = jb.JijModelingTarget(
    problem=gen_model1(),
    instance=[("model1_case1", {"d1": 1}), ("model1_case2", {"d1": 2})]
    )

t2 = jb.JijModelingTarget(
    problem=gen_model2(),
    instance=("model2_case1", {"d2": 10})
)

# t3 = jb.PyQUBOTarget()
# t4 = jb.QUBOTarget(qubo=qubo, bias=bias)

# 他にPyQUBOTarget, QUBOTargetがある。

bench = jb.Benchmark(
    {"num_reads": [3, 4, 5], "num_sweeps": [10], "x": [100, 200]},
    solver=[solver1, solver2],
    targets=[t1, t2]
)
bench.run()
```
