```python
import sys 
sys.path.append("../..")
import numpy as np
from timeseries import TimeSeries

x1 = np.arange(-10, 11, 1)
y1 = x1 + 1

x2 = np.arange(-10, 11, 1)
y2 = x2 ** 2

timeseries = TimeSeries()
timeseries.add_data("linear", x1, y1)
timeseries.add_data("quadratic", x2, y2)
timeseries.show(
    figsize=(10, 5),
    title="linear and quadratic",
    color_list=["red", "#4daf4a"],
    alpha_list=[0.3, 0.8],
    linestyle_list=["dashdot", "dashed"],
    marker_list=["v", "d"],
    xlabel="x",
    ylabel="y",
    xticks=[-10, 0, 10],
    yticks=[0, 50, 100],
)
```
